```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_weather = pd.read_csv('data/weather.csv')
df_weather.head()
```

```python
df_weather.info()
```

```python
df_weather['datetime'] = pd.date_range('2008-01-02',periods=35040,freq='H')
#df_weather["datetime"] = pd.date_range(start='2008-01-02 00:00:00', end='2011-12-31', freq='h')
df_weather.head()
```

```python
df_weather.tail()
```

```python
# Extract additional features from timestamp column
df_weather = df_weather.assign(
    timestamp = df_weather['datetime'], 
    date = lambda x: x['timestamp'].dt.date,
    year = lambda x: x['timestamp'].dt.year,
    month = lambda x: x['timestamp'].dt.month,
    day = lambda x: x['timestamp'].dt.day,
    dayofyear = lambda x: x['timestamp'].dt.dayofyear,
    hour = lambda x: x['timestamp'].dt.hour,
)
```

```python
df_weather.head()
```

```python
df_weather.columns
```

```python
sns.pairplot(data = df_weather[['Outdoor Drybulb Temperature [C]', 'Outdoor Relative Humidity [%]',
       'Diffuse Solar Radiation [W/m2]', 'Direct Solar Radiation [W/m2]','year']], 
       hue='year'
       )
```

```python
# plotting hourly temperature data over dayofyear, classified by year
plt.figure(figsize=(20,5))
sns.lineplot(data=df_weather,x='dayofyear',y='Outdoor Drybulb Temperature [C]',hue='year',ci=None)
```

```python
#smoothing the outdoor temperature data with rolling mean (unif), with centered window, window size: 1 month
# creating temperature data frame
df_temp = df_weather['Outdoor Drybulb Temperature [C]']
temp_mean = df_temp.rolling(30*24,center=True).mean()
temp_mean.tail()
```

```python
# plot of smoothed temperature data, classified by year
plt.figure(figsize=(20,5))
sns.lineplot(x=df_weather['dayofyear'],y=temp_mean,hue=df_weather['year'],ci=None)
```

- In the first 60 days of the year, bigger variance of mean temperature between years: maybe difficult to model precisely.
- 2011 (aka 2021) had the coldest summer and the mildest winter.
